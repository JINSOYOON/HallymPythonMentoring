## while, for
#### while
- 반복해서 문장을 수행해야 할 때 while문을 사용한다.
- while 문의 기본 구조
- 조건문이 있어야하고 조건문을 종료시키기 위해 조건문의 값을 조절하는 연산이 있어야한다.  

while 조건문:  
&nbsp;&nbsp;&nbsp;&nbsp;<수행할 문장1>  
&nbsp;&nbsp;&nbsp;&nbsp;<수행할 문장2>  
&nbsp;&nbsp;&nbsp;&nbsp;<수행할 문장3>  
- while 문은 조건문이 참인 동안에 while문 안에 속하는 문장들이 반복해서 수행된다. 

#### "열 번 찍어 안 넘어가는 나무 없다" 라는 속담을 파이썬으로 구현해보자

```python
name="최정명"
age = 25
weight = 80
height = 160
print("제 이름은",name,"이고 나이는",age,"이고 몸무게는",weight,"입니다")
print("제 이름은 {}이고 나이는 {}이고 몸무게는 {}입니다.".format(name,age,weight))
```

- while의 영어 단어 의미는 "~동안에" 라는 의미이다.
- 조건이 만족하는 동안에 while문이 계속해서 실행된다는 의미.
- 위 조건의 경우 나무를 10번 찍기 전까지 계속 while 문이 실행된다.

```python
treeHit = 0
while treeHit<10:
    treeHit = treeHit+1
    print("나무를 {}번 찍었습니다.".format(treeHit))
    if treeHit == 10:
        print("나무가 넘어갑니다.")
```

- 위 예에서 조건문은 treeHit<10 이다. treeHit이 10보다 작은 동안 while문 안의 문장들을 계속 수행한다.
- while문 안의 문장을 보면 treeHit=treeHit+1로 treeHit 값이 계속 1씩 증가한다.
- 나무를 treeHit 번만큼 찍었다는 문장을 출력하고 treeHit이 10이 되면 "나무가 넘어간다" 는 문장을 출력한다.
- treeHit이 10이 되면 조건문이 거짓이 되므로 while문을 빠져나가게 된다.

#### for문의 응용
- "총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 출력하시오."

```python
scores = [90, 25, 67, 45, 80]

number = 0
for score in scores:
    number = number +1
    if score >= 60:
        print("{}번 학생은 합격입니다.".format(number))
    else:
        print("{}번 학생은 불합격입니다.".format(number))
```

- 앞서 살펴보았던 60점 이상이면 합격이라는 문장을 출력하는 예제를 range함수를 이용해서 다음과 같이 바꿀 수 있다.


```python
scores = [90, 25, 67, 45, 80]
for number in range(len(scores)):
    if scores[number] < 60:
        continue
    print("{}번 학생 축하합니다. 합격입니다.".format(number+1))
```

#### for 와 range를 이용한 구구단 출력


```python
for i in range(2,10):
    for j in range(1,10):
        print("{} * {} = {}".format(i,j,i*j))
    print()
```

#### 리스트 안에 for문 포함하기


```python
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
    
print(result)
```

- 위의 코드를 아래처럼 간단하게 구현할 수 있다.

```python
a=[1,2,3,4]
result=[num*3 for num in a]
print(result)
```

#### 만약 [1,2,3,4] 중에서 짝수인 2와 4에만 3을 곱하여 담고 싶다면?
- [표현식 for 항목 in 리스트 if 조건문]
- 이를 ListComprehension 이라고 부른다.

```python
a = [1,2,3,4]
result = [num*3 for num in a if num%2==0]
print(result)
```
